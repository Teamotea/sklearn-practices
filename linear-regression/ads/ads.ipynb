## 解答

```python
import pandas as pd
```

```python
df = pd.read_csv('Advertising.csv')
df.tail()
```

```python
y = df['Sales']
X = df[['TV', 'Radio', 'Newspaper']]
X.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X, y)
yhat = reg.predict(X)
```

```python
print('y切片= ', reg.intercept_)
print('係数 = ', reg.coef_)
```

```python
SSE = ((yhat - y) ** 2).sum()
SST = ((y.mean() - y) ** 2).sum()
print('R2 =', 1 - SSE / SST)
```

```python
print(reg.score(X, y))
```

### 可視化

```python
from yellowbrick.regressor import PredictionError
```

```python
visualizer = PredictionError(reg)

visualizer.fit(X, y)
visualizer.score(X, y)
visualizer.show()
```

```python
from yellowbrick.regressor import ResidualsPlot
```

```python
visualizer = ResidualsPlot(reg)

visualizer.fit(X, y)
visualizer.score(X, y)
visualizer.show()
```
